```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
df = pd.read_csv(r'C:\Users\adria\OneDrive\Desktop\archive\players_22.csv')

# Ver as primeiras linhas
df.head()
```

```python
# Tamanho do dataset
print(f'Linhas: {df.shape[0]}')
print(f'Colunas: {df.shape[1]}')
print(f'\nColunas disponíveis:')
print(df.columns.tolist())
```

```python
# Selecionar colunas relevantes
cols = ['short_name', 'long_name', 'age', 'nationality_name', 
        'club_name', 'league_name', 'overall', 'potential',
        'value_eur', 'wage_eur', 'player_positions',
        'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']

df = df[cols]

# Remover linhas vazias
df = df.dropna(subset=['overall', 'nationality_name', 'club_name'])

print(f'Dataset limpo: {df.shape[0]} jogadores')
df.head()
```

```python
# Top 10 jogadores
top10 = df.nlargest(10, 'overall')[['short_name', 'overall']]

plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x='overall', y='short_name', palette='magma')
plt.title('Top 10 FIFA 22 Players by Overall Rating', fontsize=16)
plt.xlabel('Overall Rating')
plt.ylabel('Player')
plt.xlim(88, 94)
for i, v in enumerate(top10['overall']):
    plt.text(v + 0.05, i, str(v), va='center', fontsize=11)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(df['age'], bins=30, kde=True, color='purple')
plt.title('Age Distribution of FIFA 22 Players', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.axvline(df['age'].mean(), color='red', linestyle='--', label=f'Mean Age: {df["age"].mean():.1f}')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
top_countries = df['nationality_name'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries with Most FIFA 22 Players', fontsize=16)
plt.xlabel('Number of Players')
plt.ylabel('Country')
for i, v in enumerate(top_countries.values):
    plt.text(v + 5, i, str(v), va='center', fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# Mapa de correlação
skills = ['overall', 'potential', 'pace', 'shooting', 
          'passing', 'dribbling', 'defending', 'physic']

plt.figure(figsize=(10, 8))
sns.heatmap(df[skills].corr(), 
            annot=True, 
            fmt='.2f', 
            cmap='coolwarm',
            center=0)
plt.title('Correlation Between Player Skills — FIFA 22', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
from scipy import stats

# Definir jogadores elite (overall >= 85)
df['is_elite'] = (df['overall'] >= 85).astype(int)

# Distribuição normal do overall
mu, std = stats.norm.fit(df['overall'])

plt.figure(figsize=(12, 5))
x = range(40, 100)
p = stats.norm.pdf(x, mu, std)

sns.histplot(df['overall'], bins=40, stat='density', color='steelblue', alpha=0.6, label='Players')
plt.plot(x, p, 'r-', linewidth=2, label=f'Normal Distribution (μ={mu:.1f}, σ={std:.1f})')
plt.axvline(85, color='green', linestyle='--', linewidth=2, label='Elite threshold (85+)')
plt.title('Overall Rating Distribution & Probability of Being Elite', fontsize=16)
plt.xlabel('Overall Rating')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()

elite_prob = (df['overall'] >= 85).mean() * 100
print(f'Probability of a player being Elite (85+): {elite_prob:.2f}%')
```

```python
top_clubs = df.groupby('club_name')['overall'].mean().nlargest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=top_clubs, x='overall', y='club_name', palette='rocket')
plt.title('Top 10 Clubs by Average Overall Rating — FIFA 22', fontsize=16)
plt.xlabel('Average Overall Rating')
plt.ylabel('Club')
plt.xlim(78, 88)
for i, v in enumerate(top_clubs['overall']):
    plt.text(v + 0.05, i, f'{v:.1f}', va='center', fontsize=11)
plt.tight_layout()
plt.show()
```
